```python
class Item2Vec():

    def __init__(self):

        model = NeuMF_maskAware_sep_emb_leaky.get_model(layers, reg_layers, reg_mf)
        weights_path = pre + "4_1_leaky_relu_low_reg_weights.h5"
        model.load_weights(weights_path)

        layer_name = 'mf_item'
        self.intermediate_layer_mfitem = Model(inputs=model.input, outputs=model.get_layer(layer_name).output)
        layer_name_wg = 'mlp_embedding_item_wg'
        self.intermediate_layer_mlpwg = Model(inputs=model.input, outputs=model.get_layer(layer_name_wg).output)
        layer_name_final = 'layer3'
        self.intermediate_layer_final = Model(inputs=model.input, outputs=model.get_layer(layer_name_final).output)
        layer_name_anbieter_mkt = 'mlp_embedding_item_mkt'
        self.intermediate_layer_amkt = Model(inputs=model.input, outputs=model.get_layer(layer_name_anbieter_mkt).output)
        layer_name_anbieter = 'mlp_embedding_item_anbieter'
        self.intermediate_layer_a = Model(inputs=model.input, outputs=model.get_layer(layer_name_anbieter).output)



    def get_vec_item(self, data):
        """ Gets a vector embedding from an image
        :param image_path: path to image on filesystem
        :returns: numpy ndarray
        """

        x = Item2Vec.transform(data, 200)
        intermediate_output = self.intermediate_layer_mfitem.predict(x)
        
        return intermediate_output[0]
    
    
    def get_vec_wg(self, data):
        """ Gets a vector embedding from an image
        :param image_path: path to image on filesystem
        :returns: numpy ndarray
        """

        x = Item2Vec.transform(data, 200)
        intermediate_output = self.intermediate_layer_mlpwg.predict(x)
        
        return intermediate_output[0]
    
    def get_vec_anbieter_mkt(self, data):
        """ Gets a vector embedding from an image
        :param image_path: path to image on filesystem
        :returns: numpy ndarray
        """

        x = Item2Vec.transform(data, 200)
        intermediate_output = self.intermediate_layer_amkt.predict(x)
        
        return intermediate_output[0]
    
    def get_vec_anbieter(self, data):
        """ Gets a vector embedding from an image
        :param image_path: path to image on filesystem
        :returns: numpy ndarray
        """

        x = Item2Vec.transform(data, 200)
        intermediate_output = self.intermediate_layer_a.predict(x)
        
        return intermediate_output[0]
    
    
    def get_vec_final(self, data):
        """ Gets a vector embedding from an image
        :param image_path: path to image on filesystem
        :returns: numpy ndarray
        """

        x = Item2Vec.transform(data, 200)
        intermediate_output = self.intermediate_layer_final.predict(x)
        
        return intermediate_output[0]
    
    
    def transform(data_bundle, max_length):
        data_bundle = data_bundle.dropna().reset_index()
        # context
        month = np.nan_to_num(data_bundle.month_enc.values.reshape(-1,1))
        days_online = np.nan_to_num(data_bundle.days_online_std.values.reshape(-1,1))#, dtype = "float32")
        # item
        item_anbieter = np.nan_to_num(data_bundle.anbieterid_enc.values.reshape(-1,1)) # nur weil es noch Nan gab - fix ich noch
        item_mkt= np.nan_to_num(data_bundle.anbietermarktplatz_enc.values.reshape(-1,1))
        item_wg = np.nan_to_num(data_bundle.warengruppe_enc.values.reshape(-1,1))
        item_preis = np.nan_to_num(data_bundle.preis_std.values.reshape(-1,1))
        item_ve = np.nan_to_num(data_bundle.minve_std.values.reshape(-1,1))
        list_text = []
        list_text_user = []
        for i in range(len(data_bundle)):
            list_text.append(data_bundle.text_vec[i])
            list_text_user.append(data_bundle.text_vec_user[i])
        item_text = np.array(list_text, ndmin = 2)
        # user
        user_text = np.array(list_text_user, ndmin = 2)

        user_mkt = np.nan_to_num(data_bundle.usermkt_enc.values.reshape(-1,1))
        user_anbieter = pad_sequences(data_bundle.anbieterid_enc_user, maxlen = max_length, padding = "pre")
        user_anbietermkt = pad_sequences(data_bundle.anbietermarktplatz_enc_user, maxlen = max_length, padding = "pre")
        user_wg = pad_sequences(data_bundle.warengruppe_enc_user, maxlen = max_length, padding = "pre")
        user_preis = np.nan_to_num(data_bundle.preis_std_user.values.reshape(-1,1))
        user_ve = np.nan_to_num(data_bundle.minve_std_user.values.reshape(-1,1))
        # features 
        x_values = [month, days_online, item_anbieter, item_mkt, item_wg, item_preis, item_ve, user_text , user_mkt, user_anbietermkt, user_wg, user_anbieter, user_preis, user_ve, item_text]
        return(x_values)
```
